## nodes

```python
import sqlite3
import pandas as pd

with sqlite3.connect('./data/bach.db') as conn:
    df = pd.read_sql("SELECT * FROM nodes", conn)
```

```python
nodes = [{"id": idx+1, "label": e} for idx, e in enumerate(df["key"])]
nodes
```

```python
labels = {}
for n in nodes:
    labels[n["label"]] = n["id"]

labels
```

## edges

```python
import sqlite3
import pandas as pd

with sqlite3.connect('./data/bach.db') as conn:
    df = pd.read_sql("SELECT * FROM edges", conn)
```

```python
df.head()
```

```python
edges = [{"from": labels[e["source"]], "to": labels[e["target"]]} for e in df.to_dict("records")]
edges
```

```python
import json

with open("./data/bach_network.js", "w") as f:
    f.write(
        f"""
        const nodes_json = {json.dumps(nodes)};

        const edges_json = {json.dumps(edges)};
        """
    )
```
